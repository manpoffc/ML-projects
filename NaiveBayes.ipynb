```python
desc=[]
for value in range(len(description)):


  new_desc = re.sub(r'[^\w\s]', ' ', str(description[value]))
  new_desc = re.sub('[^a-zA-Z]',' ',str(description[value]))
  new_desc= re.sub(r'\s+[a-zA-Z]\s+', ' ', new_desc)
  new_desc=re.sub('[^A-Za-z0-9]+', ' ', new_desc)
  new_desc = re.sub(r'\s+', ' ', new_desc, flags=re.I)
  new_desc = re.sub(r'\d+', ' ', new_desc)
  new_desc= new_desc.lower()
  desc.append(new_desc)


```
